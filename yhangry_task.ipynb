## Question 2: What are the factors impacting high or low conversion?

```python
# Data cleaning
def clean_data(df):
    # Remove rows with zero or null values in columns that may be problematic later
    df = df[df['Number Of People'] > 0]  # Remove zero party size
    df = df[df['Total # Of Chefs Invited'] > 0]  # Remove zero total chefs
    return df
    
```

```python
cleaned_data = clean_data(chef_data)
```

```python
#Seeing how many entries we lost after the cleaning

print(f"Rows before cleaning: {len(chef_data)}")
print(f"Rows after cleaning: {len(cleaned_data)}")

data_removed = chef_data.shape[0] - cleaned_data.shape[0] 
perc_removed_data = data_removed/chef_data.shape[0]*100
print((f"Data removed: {perc_removed_data:.2f}%"))
```

```python
#adding extra metrics that will be useful later
def enrich_data(df):
    # Time-based metric: how long between meal date and quote creation
    df['Lead Time Days'] = (pd.to_datetime(df['Meal Date']) - 
                           pd.to_datetime(df['Quote Created At Timestamp'])).dt.days
    
    # Engagement metric: is the communication balanced between chefs and customers?
    df['Message Ratio'] = df['# Of Chef Messages'] / df['# Of Customer Messages'].replace(0, 1)
    
    return df
```

```python
#return the dataset to work with 
cleaned_data = enrich_data(cleaned_data)
```

```python
#analysing factors to see if anything interesting appears initially
def analyze_conversion_factors(df):
    
    metrics = [
        # Original metrics from spreadsheet
        'Budget', 'Number Of People', '# Chef Messages', 
        '# Of Customer Messages', 'Time to Third Quote Response (Mins)',
        'Time to Fifth Quote Response (Mins)',
        'Avg Invited Chef Travel Distance',
        'Total # of Chefs Invited',
        '# Chef Quoted',
        'Budget PP',
        
        # Derived time metrics
        'Lead Time Days',
        
        # Derived engagement metrics
        'Message Ratio',
       
    ]
    #creating an empty dataframe which will be filled with all the metrics as shown in converted & non converted cases
    comparison = pd.DataFrame()
    for metric in metrics:
        if metric in df.columns:
            # For quote response times, only use non-empty values
            # If some chefs do not respond with a quote, that's problematic in calculations and skews results
            if metric in ['Time to Third Quote Response (Mins)', 'Time to Fifth Quote Response (Mins)']:
                valid_data = df[df[metric].notna()]
                converted = valid_data[valid_data['Bookings ID'].notna()][metric].mean()
                not_converted = valid_data[valid_data['Bookings ID'].isna()][metric].mean()
            else:
                converted = df[df['Bookings ID'].notna()][metric].mean()
                not_converted = df[df['Bookings ID'].isna()][metric].mean()
            
            # Calculate difference % only if both values are valid
            if pd.notna(converted) and pd.notna(not_converted) and not_converted != 0:
                diff_pct = ((converted - not_converted) / not_converted * 100)
            else:
                diff_pct = np.nan
            
            comparison = pd.concat([comparison, pd.DataFrame({
                'Metric': [metric],
                'Converted': [converted],
                'Not Converted': [not_converted],
                'Difference %': [diff_pct]
            })])
    
    print("\nMetrics Comparison:")
    display(comparison.round(2))
```

```python
#visualising the results
analyze_conversion_factors(cleaned_data)
```

```python
#analysing non-numerical factors

def analyze_categorical(column):
        if column in cleaned_data.columns:
            conv_rates = cleaned_data.groupby(column).agg({
                'Request ID': 'count',
                'Bookings ID': lambda x: x.notna().sum()
            })
            conv_rates['Conversion Rate %'] = (conv_rates['Bookings ID'] / conv_rates['Request ID'] * 100)
            return conv_rates.sort_values('Conversion Rate %', ascending=False)
        return None
```

```python
# Analyze day of week impact
dow_conversion = analyze_categorical('Meal day of the week')
    
# Analyze serving type impact
serving_conversion = analyze_categorical('Serving Type')
    
print("\nDay of Week Conversion Rates:")
display(dow_conversion.round(2))
    
print("\nServing Type Conversion Rates:")
display(serving_conversion.round(2))
    
```

## Q4. What’s the most relevant county with the lowest conversion rate? How would you explain this? Relevant meaning >100 requests.


```python
# Calculate county-level metrics
def analyze_county_performance(df, min_requests=100):
    # Group by county and calculate metrics
    county_metrics = df.groupby('County').agg({
        'Request ID': 'count',
        'Bookings ID': lambda x: x.notna().sum(),
        'Total # Of Chefs Invited': 'mean',
        '# Chef Quoted': 'mean',
        'Avg Invited Chef Travel Distance': 'mean',
        'Time to Third Quote Response (Mins)': 'mean',
        'Budget PP': 'mean'
    }).reset_index()
    
    # Calculate conversion rate 
    county_metrics['Conversion Rate'] = (county_metrics['Bookings ID'] / county_metrics['Request ID'] * 100)
    
    # Filter for relevant counties as per the rule provided
    relevant_counties = county_metrics[county_metrics['Request ID'] >= min_requests]
    
    return relevant_counties


county_performance = analyze_county_performance(cleaned_data)

# Sort by conversion rate to see top counties
print("\nTop 5 Counties by Conversion Rate (>100 requests):")
display(county_performance.sort_values('Conversion Rate', ascending=False)
        .head().round(2))

# Sort by conversion rate to see bottom counties
print("\nBottom 5 Counties by Conversion Rate (>100 requests):")
display(county_performance.sort_values('Conversion Rate', ascending=False)
        .tail().round(2))

```

## Q6. Which 3 counties are most in need of more chefs and why?

```python
def analyze_chef_needs(df, min_requests=100):
    """
    Analyze which counties need more chefs based on key performance metrics
    """
    #Group data by county and calculate key metrics
    county_metrics = df.groupby('County').agg({
        # Count total requests per county
        'Request ID': 'count',
        # Count successful bookings (non-null Bookings ID)
        'Bookings ID': lambda x: x.notna().sum(),
        # Sum all chefs invited
        'Total # Of Chefs Invited': 'sum',
        # Sum all quotes received
        '# Chef Quoted': 'sum',
        # Calculate average travel distance
        'Avg Invited Chef Travel Distance': 'mean',
        # Calculate average response time
        'Time to Third Quote Response (Mins)': 'mean',
        # Calculate average budget
        'Budget': 'mean'
    }).reset_index()
    
    # Rename columns for clarity
    county_metrics.columns = [
        'County', 'Total_Requests', 'Total_Bookings', 
        'Total_Chefs_Invited', 'Total_Quotes', 
        'Avg_Travel_Distance', 'Avg_Response_Time',
        'Avg_Budget'
    ]
    
    #: Calculate key performance indicators
    county_metrics['Conversion_Rate'] = (county_metrics['Total_Bookings'] / 
                                       county_metrics['Total_Requests'])
    
    county_metrics['Quote_Rate'] = (county_metrics['Total_Quotes'] / 
                                   county_metrics['Total_Chefs_Invited'])
    
    county_metrics['Quotes_Per_Request'] = (county_metrics['Total_Quotes'] / 
                                          county_metrics['Total_Requests'])
    
    # Filter for relevant counties (>100 requests)
    relevant_counties = county_metrics[county_metrics['Total_Requests'] >= min_requests].copy()
    
    # Calculate 'chef need' score
    # Higher score = greater need for chefs

    # Components:
    
    # Low conversion rate indicates unmet demand (1 - conversion_rate)
    # High travel distance indicates poor coverage (normalized by max distance)
    # Low quote rate indicates chef shortage (1 - quote_rate)
    # High request volume indicates market importance (normalized by max requests)
    
    max_travel = relevant_counties['Avg_Travel_Distance'].max()
    max_requests = relevant_counties['Total_Requests'].max()
    
    #adding weights to all the components 
    
    relevant_counties['Chef_Need_Score'] = (
        # Demand factor: Higher unmet demand = higher need
        (1 - relevant_counties['Conversion_Rate']) * 0.3 +
        # Coverage factor: Higher travel distance = higher need
        (relevant_counties['Avg_Travel_Distance'] / max_travel) * 0.25 +
        # Supply factor: Lower quote rate = higher need
        (1 - relevant_counties['Quote_Rate']) * 0.25 +
        # Market size factor: Higher request volume = higher need
        (relevant_counties['Total_Requests'] / max_requests) * 0.2
    )
    
    return relevant_counties.sort_values('Chef_Need_Score', ascending=False)

# Run the analysis

county_needs = analyze_chef_needs(cleaned_data)

# Display results
print("\nTop 10 Counties Most in Need of Chefs:")
columns_to_show = [
    'County', 'Total_Requests', 'Conversion_Rate', 
    'Quote_Rate', 'Avg_Travel_Distance', 'Chef_Need_Score'
]
display(county_needs[columns_to_show].head(10).round(3))

# Detailed analysis of top 3 counties
print("\nDetailed Analysis of Top 3 Counties:")
for _, county in county_needs.head(3).iterrows():
    print(f"\n{county['County']}:")
    print(f"Total Requests: {int(county['Total_Requests'])}")
    print(f"Conversion Rate: {county['Conversion_Rate']*100:.1f}%")
    print(f"Quote Rate: {county['Quote_Rate']*100:.1f}%")
    print(f"Quotes Per Request: {county['Quotes_Per_Request']:.1f}")
    print(f"Average Travel Distance: {county['Avg_Travel_Distance']:.1f}km")
    print(f"Average Response Time: {county['Avg_Response_Time']/60:.1f} hours")
```
